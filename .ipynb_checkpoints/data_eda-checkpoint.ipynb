```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_clean.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace = True)
```

```python
pd.set_option('display.max_columns', None)
df.head(20)
```

```python
df.columns
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
# Simplify title
df['Job_simp'] = df['Job Title'].apply(title_simplifier)
df['Job_simp'].value_counts()
```

```python
# Simplify title
df['seniority'] = df['Job Title'].apply(seniority)
df['seniority'].value_counts()
```

```python
# Description length
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
df['desc_len']
```

```python
# Competitor count
df['Competitors'] = df['Competitors'].apply(lambda x: len(x.split(',')) if x != '-1' else x)
```

```python
df['Competitors'].value_counts()
```

```python
# Turn salary per hour to salary per year
# ~2000 hours per year 
df['min_salary'] = df.apply(lambda x: (x.min_salary)*2 if x.hourly == 1 else x.min_salary, axis = 1)
df['max_salary'] = df.apply(lambda x: (x.max_salary)*2 if x.hourly == 1 else x.max_salary, axis = 1)
```

```python
pd.set_option('display.max_rows', None)
df[['hourly', 'min_salary', 'max_salary']]
```

```python
df['Job Description'] = df['Job Description'].apply(lambda x: x.replace('\n', ' '))
```
